```python
import random

list_1 = []
prize_1 = []
for i in range(5):
    user_value = input("What is your name? ")
    list_1.append(user_value)
for i in range(3):
    prize_value = input("Enter prize (Grand Prize should be last): ")
    prize_1.append(prize_value)

winners = random.sample(list_1, 3)
prizes = prize_1  
for index, (person, prize) in enumerate(zip(winners, prizes), start=1):
    if index == 3:
        print(f"{person} is the THIRD winner! ðŸŽ‰ Grand prize: {prize}!")
    else:
        print(f"{person} is a winner and wins: {prize}!")
```
